### Transformation des Variables CatÃ©goriques

Nous allons transformer :

Education,
Marital_Status

```python
from sklearn.preprocessing import OneHotEncoder

# Encodage des variables catÃ©goriques
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_categorical = encoder.fit_transform(df[["Education", "Marital_Status"]])
encoded_df = pd.DataFrame(encoded_categorical, columns=encoder.get_feature_names_out(["Education", "Marital_Status"]))

# Fusionner avec le dataset initial (sans les colonnes d'origine)
df = pd.concat([df.drop(columns=["Education", "Marital_Status"]), encoded_df], axis=1)
```

### CrÃ©ation de Nouvelles Variables Pertinentes

```python
# CrÃ©ation des nouvelles variables
df["Total_Spending"] = df[["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]].sum(axis=1)
df["Total_Purchases"] = df[["NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases"]].sum(axis=1)

# Score d'engagement = (nombre d'achats) / (rÃ©cence dâ€™achat)
df["Engagement_Score"] = df["Total_Purchases"] / (df["Recency"] + 1)  # Ajout de +1 pour Ã©viter la division par 0
```

### Normalisation des Variables NumÃ©riques

```python
from sklearn.preprocessing import StandardScaler

# Liste des colonnes Ã  normaliser
cols_to_scale = ["Income", "Total_Spending", "Total_Purchases", "Engagement_Score", "Recency"]

scaler = StandardScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

# Affichage des donnÃ©es prÃªtes pour la segmentation
print(df.head())
```

## Segmentation des Clients avec le Clustering

### DÃ©termination du Nombre Optimal de Clusters

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# SÃ©lection des variables pour le clustering
features = ["Income", "Total_Spending", "Total_Purchases", "Engagement_Score", "Recency"]
X = df[features]

# Tester plusieurs nombres de clusters
inertia = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# Tracer la courbe du coude (Elbow Method)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(K_range, inertia, marker='o')
plt.xlabel("Nombre de Clusters")
plt.ylabel("Inertie")
plt.title("MÃ©thode du Coude")

# Tracer le score de silhouette
plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette_scores, marker='o', color='red')
plt.xlabel("Nombre de Clusters")
plt.ylabel("Score de Silhouette")
plt.title("Score de Silhouette")
plt.show()
```

D'aprÃ¨s tes graphiques, voici les observations :

MÃ©thode du Coude 

L'inertie diminue fortement entre 2 et 4 clusters, puis la baisse devient plus progressive.
Le coude semble se situer autour de 3 ou 4 clusters.
Score de Silhouette 

Le score est maximal pour k = 3 (â‰ˆ 0.39) et chute brutalement aprÃ¨s.
Pour k = 4, le score est encore acceptable mais commence Ã  se stabiliser.
Conclusion : On choisit k = 3
Ce choix est justifiÃ© car :
La mÃ©thode du coude suggÃ¨re un ralentissement de la diminution Ã  k=3.
Le score de silhouette est maximal pour k=3.

### DBSCAN

##### MÃ©thode du coude

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Choisir le mÃªme nombre que `min_samples`
k = 10  

# Calcul des distances entre chaque point et ses k voisins les plus proches
nearest_neighbors = NearestNeighbors(n_neighbors=k)
nearest_neighbors.fit(X)
distances, indices = nearest_neighbors.kneighbors(X)



# Trier les distances du plus petit au plus grand
distances = np.sort(distances[:, k-1])

# Tracer le graphe des distances
plt.plot(distances)
plt.xlabel("Points du dataset triÃ©s")
plt.ylabel(f"Distance au {k}-Ã¨me plus proche voisin")
plt.title("MÃ©thode du Coude pour eps (DBSCAN)")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# EntraÃ®nement du modÃ¨le DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=15)  
clusters_dbscan = dbscan.fit_predict(X)

# Ajout des labels au DataFrame
df["DBSCAN_Cluster"] = clusters_dbscan

if isinstance(X, pd.DataFrame):
    X = X.to_numpy()

# Visualisation des clusters DBSCAN
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=clusters_dbscan, cmap="viridis", alpha=0.6)
plt.title("Segmentation des Clients avec DBSCAN")
plt.show()
```

### Clustering HiÃ©rarchique (Agglomerative)

```python
from sklearn.cluster import AgglomerativeClustering

# EntraÃ®nement du modÃ¨le Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=optimal_clusters)
clusters_agglo = agglo.fit_predict(X)

# Ajout des labels au DataFrame
df["Agglo_Cluster"] = clusters_agglo

if isinstance(X, pd.DataFrame):
    X = X.to_numpy()

# Visualisation des clusters Agglomerative Clustering
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=clusters_agglo, cmap="coolwarm", alpha=0.6)
plt.title("Segmentation des Clients avec Clustering HiÃ©rarchique")
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des modÃ¨les de clustering
models = {
    "K-Means": "KMeans_Cluster",
    "DBSCAN": "DBSCAN_Cluster",
    "Agglomerative Clustering": "Agglo_Cluster"
}

# Calcul des scores de silhouette et de Davies-Bouldin
scores = {}
for model_name, column in models.items():
    if column in df.columns:
        labels = df[column]
        if len(set(labels)) > 1:  # Ã‰viter les erreurs avec DBSCAN (-1)
            silhouette = silhouette_score(X, labels)
            davies_bouldin = davies_bouldin_score(X, labels)
            scores[model_name] = {"Silhouette": silhouette, "Davies-Bouldin": davies_bouldin}

# Affichage des rÃ©sultats
score_df = pd.DataFrame(scores).T
print("\n Scores des ModÃ¨les de Clustering :")
print(score_df)

#  Visualisation des clusters avec ACP (Analyse en Composantes Principales)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, (model_name, column) in enumerate(models.items()):
    if column in df.columns:
        sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df[column], palette="coolwarm", ax=axes[i])
        axes[i].set_title(f"Visualisation des Clusters ({model_name})")
        axes[i].set_xlabel("Composante Principale 1")
        axes[i].set_ylabel("Composante Principale 2")

plt.show()
```

### Analyse en Composantes Principales (ACP - PCA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# SÃ©lection des variables utilisÃ©es pour le clustering
features = ["Income", "Total_Spending", "Total_Purchases", "Engagement_Score", "Recency"]
X = df[features]

# Appliquer l'ACP pour rÃ©duire Ã  2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Ajouter les composantes principales dans le DataFrame
df["PCA_1"] = X_pca[:, 0]
df["PCA_2"] = X_pca[:, 1]

# Visualisation des clusters avec PCA
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["PCA_1"], y=df["PCA_2"], hue=df["Agglo_Cluster"], palette="coolwarm", alpha=0.7)
plt.title("Visualisation des Clusters avec ACP (PCA)")
plt.xlabel("Composante Principale 1")
plt.ylabel("Composante Principale 2")
plt.legend(title="Cluster")
plt.show()
```

Lâ€™ACP permet de rÃ©duire la dimension des donnÃ©es tout en conservant lâ€™essentiel de lâ€™information.
On observe que les clusters sont globalement bien sÃ©parÃ©s, bien que la sÃ©paration ne soit pas parfaite.
Le cluster gris (Cluster 1) semble moins dÃ©fini que les deux autres.
Cela signifie que certaines variables influencent les clusters d'une maniÃ¨re non linÃ©aire.

### t-SNE (t-Distributed Stochastic Neighbor Embedding)

```python
from sklearn.manifold import TSNE

# Appliquer t-SNE sur les donnÃ©es
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

# Ajouter les coordonnÃ©es t-SNE dans le DataFrame
df["tSNE_1"] = X_tsne[:, 0]
df["tSNE_2"] = X_tsne[:, 1]

# Visualisation des clusters avec t-SNE
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["tSNE_1"], y=df["tSNE_2"], hue=df["Agglo_Cluster"], palette="coolwarm", alpha=0.7)
plt.title("Visualisation des Clusters avec t-SNE")
plt.xlabel("t-SNE Composante 1")
plt.ylabel("t-SNE Composante 2")
plt.legend(title="Cluster")
plt.show()
```

Contrairement Ã  lâ€™ACP, le t-SNE est une technique de rÃ©duction de dimension qui conserve mieux les structures locales.
On voit clairement trois groupes bien distincts.
Cela confirme que nos clusters identifiÃ©s par K-Means ont une cohÃ©rence structurelle.
