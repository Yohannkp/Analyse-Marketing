## DÃ©tection des valeurs aberrantes

```python
from scipy.stats import zscore

# DÃ©tection des valeurs aberrantes avec le Z-score pour 'Income' et 'Total_Spending'
df["Income_Zscore"] = zscore(df["Income"])
df["Spending_Zscore"] = zscore(df["Total_Spending"])

# Filtrer les valeurs aberrantes (Z-score > 3 ou < -3)
outliers_income = df[df["Income_Zscore"].abs() > 3]
outliers_spending = df[df["Spending_Zscore"].abs() > 3]

print(f"ðŸ”¹ Nombre de valeurs aberrantes dans 'Income' : {outliers_income.shape[0]}")
print(f"ðŸ”¹ Nombre de valeurs aberrantes dans 'Total_Spending' : {outliers_spending.shape[0]}")

# Afficher quelques valeurs aberrantes dÃ©tectÃ©es
display(outliers_income.head())
display(outliers_spending.head())
```

La majoritÃ© de ces clients ont un niveau d'Ã©ducation Ã©levÃ© (PhD, Master) et sont mariÃ©s ou en couple.
Ces valeurs peuvent Ãªtre vÃ©rifiÃ©es avant dâ€™Ãªtre exclues ou normalisÃ©es.

### Objectif : Identifier les segments de clients les plus rentables en fonction de leurs interactions avec les campagnes marketing

```python
# CrÃ©ation d'une variable "Total_Campaigns_Accepted"
df["Total_Campaigns_Accepted"] = (
    df["AcceptedCmp1"] + df["AcceptedCmp2"] + df["AcceptedCmp3"] +
    df["AcceptedCmp4"] + df["AcceptedCmp5"] + df["Response"]
)

# CatÃ©gorisation des clients selon leur engagement avec les campagnes
df["Engagement_Campagnes"] = pd.cut(df["Total_Campaigns_Accepted"], 
                                    bins=[0, 1, 3, 6], 
                                    labels=["Faible", "Moyen", "Ã‰levÃ©"])

# CatÃ©gorisation des clients selon leurs dÃ©penses totales
df["Segment_Spending"] = pd.qcut(df["Total_Spending"], q=3, labels=["Bas", "Moyen", "Ã‰levÃ©"])

# CatÃ©gorisation des clients selon leur revenu
df["Segment_Income"] = pd.qcut(df["Income"], q=3, labels=["Bas", "Moyen", "Ã‰levÃ©"])

# Affichage des segments
display(df[["Total_Campaigns_Accepted", "Engagement_Campagnes", "Segment_Spending", "Segment_Income"]].head())
```

Engagement avec les campagnes :

Faible : 0 ou 1 campagne acceptÃ©e.
Moyen : 2 ou 3 campagnes acceptÃ©es.
Ã‰levÃ© : 4 Ã  6 campagnes acceptÃ©es.
DÃ©penses totales (Segment_Spending) :

Bas : Clients qui dÃ©pensent peu.
Moyen : Clients qui ont une consommation moyenne.
Ã‰levÃ© : Clients qui dÃ©pensent beaucoup.
Revenu (Segment_Income) :

Bas : Revenus les plus faibles.
Moyen : Revenus intermÃ©diaires.
Ã‰levÃ© : Revenus les plus Ã©levÃ©s.
